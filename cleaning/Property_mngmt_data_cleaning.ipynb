```python
import pandas as pd
```

```python

#Function assigning a non standard level-floor to other category
def assign_level(item):
    approved_lst = ["Ισόγειο", "1ος", "2ος", "3ος", "4ος", "5ος", "6ος", "7ος", "8ος", "Υπερυψωμένο", "Ημιυπόγειο", "Ημιώροφος" ]
    try:
        if item not in approved_lst:
            item = "'Αλλο"
            return item
        else:
            return item
    except:
        pass
```

```python
#Function assigning more than 10 bedrooms to category 10
def assign_bedroom(item):
    try:
        if item >9:
            item = 10
            return item
        else:
            return item
    except:
        pass
```

```python
#Function assigning more than 10 bedrooms to category 10
def assign_bathroom(item):
    try:
        if item >5:
            item = 6
            return item
        else:
            return item
    except:
        pass
```

```python
#Flag properties that have (1) or not (0) any type of parking
df["parking_flag"] = df["parking"].notnull()
df["parking_flag"] = df["parking_flag"].replace(True, 1)
df["parking_flag"] = df["parking_flag"].replace(False, 0)
df.shape
```

```python
df = df.loc[(df["location_region"] == "Αττική")]
df.shape
```

```python
df["levels"].unique()
```

```python
df.groupby("levels").count().reset_index().sort_values("location_region", ascending=False).head(20)
```

```python
df["levels"] = list(map(assign_level, df['levels']))
df["levels"].unique()
```

```python
df.groupby("levels").count()
```

```python
df.groupby("bedrooms").count()
```

```python
df["bedrooms"] = list(map(assign_bedroom, df['bedrooms']))
df["bedrooms"].unique()
```

```python
df.groupby("bathrooms").count()
```

```python
df["bathrooms"] = list(map(assign_bathroom, df['bathrooms']))
df["bathrooms"].unique()
```
